```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
co_occurrence_mat = [[14, 7, 5, 1], 
                     [7, 12, 0, 0],
                     [5, 0, 8, 12],
                     [1, 0, 12, 2]]
```

```python
co_occurrence_mat = np.array(co_occurrence_mat)
```

```python
U, S, V = np.linalg.svd(co_occurrence_mat)
```

```python
plt.scatter(U[:, 0], U[:, 1], c=S)
plt.xlabel("SVD Component 1")
plt.ylabel("SVD Component 2")
plt.show()
```

```python
d = 2  # Number of singular values to keep (you can change this)
reconstructed_A = U[:, :d] @ np.diag(S[:d]) @ V[:d, :]

print("Original Matrix:")
print(co_occurrence_mat)
print("\nReconstructed Matrix (with d singular values):")
print(reconstructed_A)
```
